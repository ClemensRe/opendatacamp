```python
import pandas as pd
import pathlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
PROCESSED_DIR = pathlib.Path("../data/processed/")
relevant_entries = ["Bad", "West", "Lene", "Stadion"]
```

```python
for entry in relevant_entries:
    analysis_df = pd.DataFrame()
    for time_series_df_path in PROCESSED_DIR.glob("*.csv"):
        if entry in str(time_series_df_path):
            df = pd.read_csv(time_series_df_path)
            analysis_df = analysis_df.append(df)

    analysis_df = analysis_df.set_index("Date")
    analysis_df.plot.kde()
    print(f"Histogram of {entry}")
    plt.show()
```

```python
relevant_entries = ["Bad", "West", "Lene"]
```

```python
for entry in relevant_entries:
    analysis_df = pd.DataFrame()
    for time_series_df_path in PROCESSED_DIR.glob("*.csv"):
        if entry in str(time_series_df_path):
            df = pd.read_csv(time_series_df_path)
            analysis_df = analysis_df.append(df)

    analysis_df = analysis_df.set_index("Date")
    analysis_df = analysis_df.query("CountOfPersons >= 300")
    analysis_df.plot.kde(figsize=(20,10), title=f"Maximum Treshold Analysis of Entry {entry}")
    mean = analysis_df.mean()
    std = analysis_df.std()
    median = analysis_df.median()
    print(f"Histogram of {entry} with mean {mean.values}, median {median.values} and std {std.values}")
    plt.axvline(x=mean.values, linestyle='dashed', linewidth=2, color="r", label="Mean")
    plt.axvline(x=median.values, linestyle='dashed', linewidth=2, color="g", label="Median")
    plt.legend()
    plt.show()
```
