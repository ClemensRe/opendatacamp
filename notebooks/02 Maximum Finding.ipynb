```python
import pandas as pd
import pathlib
import matplotlib.pyplot as plt
```

```python
PROCESSED_DIR = pathlib.Path("../data/processed/")
relevant_entries = ["Bad", "West", "Lene"]
```

```python
for entry in relevant_entries:
    analysis_df = pd.DataFrame()
    for time_series_df_path in PROCESSED_DIR.glob("*.csv"):
        if entry in str(time_series_df_path):
            df = pd.read_csv(time_series_df_path)
            analysis_df = analysis_df.append(df)

    analysis_df = analysis_df.set_index("Date")
    analysis_df.query("CountOfPersons >= 300").hist(bins=20)
    print(f"Histogram of {entry}")
    plt.show()
```
